```python
from google.colab import drive
import os
```

```python
# Mount your Google Drive
drive.mount('/content/gdrive')
```

```python
# Specify the path to your video folder on Google Drive (replace with your actual path)
folder_path = '/content/gdrive/MyDrive/Major FINAL RUN/Final Audio Dataset'
```

```python
# Initialize video count
video_count = 0
```

```python
# Loop through files in the folder
for filename in os.listdir(folder_path):
  # Check if it's a video file
  if filename.endswith('.mp4'):
    video_count += 1
    print(filename)  # Print video name
```

```python
# Print total video count
print(f"Total Videos: {video_count}")
```
